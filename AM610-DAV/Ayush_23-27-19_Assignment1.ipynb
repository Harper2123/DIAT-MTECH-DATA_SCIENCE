```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Housepriceprediction.csv')
df.head()
```

## Q1) Write a Python code for computing the correlation coefficient for all the features with the price.

```python
corr_matrix = df.corr()
corr_with_price = corr_matrix['SalePrice']

#Tabulating the correlation coefficients
corr_table = pd.DataFrame({'Feature': corr_with_price.index, 'Correlation with Price': corr_with_price.values})

print(corr_table)
```

```python
# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data=pd.DataFrame(corr_with_price), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap: Features vs. Price')
plt.show();
```

### Inferences

From the above tabular data, or from the heatmap above, the following points can be inferred:-

- None of the features are negatively correlated to Sale Price (cannot consider ID as feature).
- LotArea show a very low positive correlation with Price (the impact of Lot Area on Sale Price is minimal).
- Total Rooms Above Ground is moderately correlated with Price (the impact cannot be ignored).
- All other variables show high correlation with Sale Price especially OverallQual and GrLivArea (highly impact the Sale Price).

## Q2) Write a Python code for Computing the correlation coefficient for all the features among themselves, and give out the inferences.

```python
corr_matrix_all = pd.DataFrame(df.corr())

corr_matrix_all
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(data=corr_matrix_all, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap: Features Among Themselves')
plt.show()
```

### Inferences
Based on the tabular data above and the heatmap, we can infer the following:-
- There is no negative correlation among variables.
- The darker shades of red are highly positively correlated variables.
- The lighter shades of red are lowly positively correlated variables.
